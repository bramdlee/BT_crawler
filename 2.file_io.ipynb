항상 인코딩을 utf8으로 설정해주면 이런 문제를 방지할 수 있습니다.코드를 수정해봅니다.

```python
# 137 예제, 파일 열기
f = open('stockcode.txt', 'rb') # 'rb' 의 경우는 binary 타입으로 파일을 읽는 다는 말이지요.
```

```python
f.read().decode('cp949') # 적절한 decode 를 붙여줘야 합니다.
```

```python
f.close() #열었으면 닫아줘야 합니다... 아니면 다른 프로세스에서 접근을 못할 수 있습니다. 특히나 쓰기중이라면요.
```

파이썬3에서 파일 입출력시 한글이던 영문이든 스페인어이든 utf8 으로 인코딩을 명시해두고 쓰거나 읽거나 해야 이러한 불상사를 막을 수 있습니다.<br>
그럼 이러한 인코딩을 UTF8 으로 바꾸어서 새로운 파일을 저장해봅시다.

```python
# 똑같이 파일을 불러옵니다.
```

```python
f = open('stockcode.txt', 'rb') 
f_text = f.read().decode('cp949')
f.close()
```

```python
print(f_text)
```

```python
new_f = open('stockcode_utf8.txt', 'w', encoding='utf8')
new_f.write(f_text) # read & write 입니다.
new_f.close()
```

항상 이렇게 파일을 열고(open) 쓰거나 읽은 다음 닫아(close) 줘야하는 건가요<br> 그래서 with라는 구문이 있습니다. entry exit 이 내포되어 있어서 알아서 열고 닫아줍니다.

```python
with open('./stockcode_utf8.txt', 'r', encoding='utf8') as f: # with 구문 내에 f 라는 변수명으로 호출 할수 있도록 as f 로 명시해줍니다.
    all_text = f.read() #이번에는 한줄이 아닌 전체로 불러서 all_text 객체에 저장하겠습니다.
```

```python
all_text # 앞에서도 보았지만 구별할 수가 없네요, string 객체로 잡혀 있는 객체는 split 을 사용 할 수 있습니다.
```

```python
type(all_text)
```

```python
all_text.split('\n') 
```

```python
all_code_list = all_text.split('\n')
type(all_code_list) #객체의 타입이 바뀝니다.
```
